## 전처리 과정에서 제거할 요인들

- 시작지점명, 도착지점명 : 중간에 도로에서의 속도가 주요할 것같아서 일단 배제한 후 훈련
- id, 날짜 : 요일이라는 요소가 따로 있기 때문에 아주 중요한 요소는 아닐거라 판단된다.
- 통과제한차량 : 통과 제한 차량이 유의미한 결론을 내지 못할 거라 판단된다.
- 도로사용여부 : 공지사항

```python
train_df2 = train_df.drop(['start_node_name','end_node_name','vehicle_restricted','base_date','id','road_name','start_latitude','start_longitude',
                            'end_latitude','end_longitude'],axis=1)
```

```python
train_df2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
test_df2 = test_df.drop(['start_node_name','end_node_name','vehicle_restricted','base_date','id','road_name','start_latitude','start_longitude',
                            'end_latitude','end_longitude'],axis=1)
```

```python
str_col = ['start_turn_restricted','end_turn_restricted']
for i in str_col:
    le = LabelEncoder()
    le=le.fit(train_df2[i])
    train_df2[i]=le.transform(train_df2[i])
    
    for label in np.unique(test_df2[i]):
        if label not in le.classes_: 
            le.classes_ = np.append(le.classes_, label)
    test_df2[i]=le.transform(test_df2[i])

pd_df = pd.get_dummies(train_df2['day_of_week'])
train_df2 = pd.concat([train_df2, pd_df], axis=1)
pd_df = pd.get_dummies(test_df2['day_of_week'])
test_df2 = pd.concat([test_df2, pd_df], axis=1)

train_df3 = train_df2.drop(['day_of_week'],axis=1)
test_df3 = test_df2.drop(['day_of_week'],axis=1)
```

```python
train_df3
```

```python
y_train = train_df3['target']
x_train = train_df3.drop(['target'],axis=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(test_df3.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error

import xgboost as xgb
import lightgbm as lgb
```

```python
import time
start_time = time.time()
total_time = start_time
model_gb = GradientBoostingRegressor(learning_rate=0.01).fit(x_train, y_train)
now = time.time()
print(f'gb 끝 : \n걸린시간 : {now-start_time:.3f/60}분' )

start_time = time.time()
model_xgb = xgb.XGBRegressor().fit(x_train, y_train)
now = time.time()
print(f'xgb 끝 : \n걸린시간 : {now-start_time:.3f/60}분' )

start_time = time.time()
model_lgb = lgb.LGBMRegressor().fit(x_train, y_train)
now = time.time()
print(f'lgb 끝 : \n걸린시간 : {now-start_time:.3f/60}분' )

start_time = time.time()
model_rf = RandomForestRegressor().fit(x_train, y_train)
now = time.time()
print(f'rf 끝 : \n걸린시간 : {now-start_time:.3f/60}분' )
print(f'fit 완료\n총 걸린시간 : {now-total_time:.3f/60}분')
```

```python
pred_gb = model_gb.predict(test_df3)
pred_xgb = model_xgb.predict(test_df3)
pred_lgb = model_lgb.predict(test_df3)
pred_rf = model_rf.predict(test_df3)

```
